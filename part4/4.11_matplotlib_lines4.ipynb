```python
#  
import pandas as pd
import matplotlib.pyplot as plt

# matplotlib     
from matplotlib import font_manager, rc
font_path = "./data/malgun.ttf"   # 
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# Excel    
df = pd.read_excel('./data/__.xlsx')

# (NaN)    (   )
df = df.ffill()

#       
mask = (df[''] == '') & (df[''] != '') 
df_seoul = df[mask]
df_seoul = df_seoul.drop([''], axis=1)
df_seoul = df_seoul.rename({'':''}, axis=1)
df_seoul = df_seoul.set_index('')

#  '','', '', ''     
col_years = list(map(str, range(1970, 2018)))
df_4 = df_seoul.loc[['','', '', ''], col_years]

#   
plt.style.use('ggplot') 

#    (figure 1   )
fig = plt.figure(figsize=(20, 10))   
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

# axe  plot   
ax1.plot(col_years, df_4.loc['',:], marker='o', markerfacecolor='green', 
        markersize=10, color='olive', linewidth=2, label=' -> ')
ax2.plot(col_years, df_4.loc['',:], marker='o', markerfacecolor='blue', 
        markersize=10, color='skyblue', linewidth=2, label=' -> ')
ax3.plot(col_years, df_4.loc['',:], marker='o', markerfacecolor='red', 
        markersize=10, color='magenta', linewidth=2, label=' -> ')
ax4.plot(col_years, df_4.loc['',:], marker='o', markerfacecolor='orange', 
        markersize=10, color='yellow', linewidth=2, label=' -> ')

#  
ax1.legend(loc='best')
ax2.legend(loc='best')
ax3.legend(loc='best')
ax4.legend(loc='best')

#   
ax1.set_title(' ->   ', size=15)
ax2.set_title(' ->   ', size=15)
ax3.set_title(' ->   ', size=15)
ax4.set_title(' ->   ', size=15)

#    
ax1.set_xticks(range(len(col_years)))
ax2.set_xticks(range(len(col_years)))
ax3.set_xticks(range(len(col_years)))
ax4.set_xticks(range(len(col_years)))

#      90 
ax1.set_xticklabels(col_years, rotation=90)
ax2.set_xticklabels(col_years, rotation=90)
ax3.set_xticklabels(col_years, rotation=90)
ax4.set_xticklabels(col_years, rotation=90)

plt.show()  #    
```
