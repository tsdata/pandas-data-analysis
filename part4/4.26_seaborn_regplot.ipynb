```python
#  
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn   
titanic = sns.load_dataset('titanic')
 
#    (5: darkgrid, whitegrid, dark, white, ticks)
sns.set_style('darkgrid')

#    (figure 2   )
fig, axes = plt.subplots(1, 2, figsize=(15, 5))   
 
#   -  (fit_reg=True)
sns.regplot(x='age',        #x 
            y='fare',       #y 
            data=titanic,   #
            ax=axes[0])     #Axes  - 1  

#   -  (fit_reg=False)
sns.regplot(x='age',        #x 
            y='fare',       #y 
            data=titanic,   #
            ax=axes[1],     #Axes  - 2         
            fit_reg=False)  # 

plt.show()
```
