```python
#  
import pandas as pd
import matplotlib.pyplot as plt

# matplotlib     
from matplotlib import font_manager, rc
font_path = "./data/malgun.ttf"   # 
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# Excel    
df = pd.read_excel('./data/__.xlsx')

# (NaN)    (   )
df = df.ffill()

#       
mask = (df[''] == '') & (df[''] != '') 
df_seoul = df[mask]
df_seoul = df_seoul.drop([''], axis=1)
df_seoul = df_seoul.rename({'':''}, axis=1)
df_seoul = df_seoul.set_index('')

#        
sr_one = df_seoul.loc['']

#   
plt.style.use('ggplot') 

#   
plt.figure(figsize=(14, 5))

# x   
plt.xticks(size=10, rotation='vertical')

# x, y  plot   
plt.plot(sr_one.index, sr_one.values, marker='o', markersize=10)  

plt.title(' ->   ', size=30)  # 
plt.xlabel('', size=20)                  #x 
plt.ylabel(' ', size=20)           #y 

# 
plt.legend(labels=[' -> '], loc='best', fontsize=15)

# y   (, )
plt.ylim(50000, 800000)

#   - 
plt.annotate('',
             xy=(20, 620000),       #  ()
             xytext=(2, 290000),    #  ()
             xycoords='data',       #
             arrowprops=dict(arrowstyle='->', color='skyblue', lw=5), # 
             )

plt.annotate('',
             xy=(47, 450000),       #  ()
             xytext=(30, 580000),   #  ()
             xycoords='data',       #
             arrowprops=dict(arrowstyle='->', color='olive', lw=5),  # 
             )

#   - 
plt.annotate(' (1970-1995)',  # 
             xy=(10, 350000),            #  
             rotation=25,                # 
             va='baseline',              #  
             ha='center',                #  
             fontsize=15,                # 
             )

plt.annotate(' (1995-2017)',  # 
             xy=(40, 470000),            #  
             rotation=-11,               # 
             va='baseline',              #  
             ha='center',                #  
             fontsize=15,                # 
             )

plt.show()  #    
```
